```python
def save_img(fn, arr):
    assert arr.max() == 1
    arr = (255*arr).astype(np.uint8)
    img = Image.fromarray(arr)
    img.save(fn)
```

```python
# init A

m = 100 
n = 784

std_dev = 1 / m
A = np.random.normal(0, std_dev, (m, n))

fn_x = 'data/mnist/img_1.jpg'
x = np.array(Image.open(fn_x)) / 255.

y = np.dot(A, x.flatten())
y = y.reshape(int(m**0.5), int(m**0.5)) # display y as square
y = (y - y.min()) / (y.max() - y.min())
print(x.shape, y.shape)
```

```python
fn_y = f'{fn_x.split(".")[0]}_y{m}.jpg'
save_img(fn_y, y)
```
